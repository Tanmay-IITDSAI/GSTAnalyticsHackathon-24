###checksum.py file contains this
import hashlib

import argparse

def calculate_checksum(file_path, hash_algorithm='sha256'):

    try:
        hash_func = hashlib.new(hash_algorithm)
        with open(file_path, 'rb') as file:
            while chunk := file.read(8192):
                hash_func.update(chunk)
        return hash_func.hexdigest()
    except FileNotFoundError:
        return f"Error: File '{file_path}' not found."
    except ValueError:
        return f"Error: Invalid hash algorithm '{hash_algorithm}'."
    except Exception as e:
        return f"Error: {str(e)}"

def main():

    parser = argparse.ArgumentParser(description="Verify the checksum of a file.")
    parser.add_argument("file_path", type=str, help="Path to the file to be checksummed.")
    parser.add_argument("-a", "--algorithm", type=str, choices=['md5', 'sha1', 'sha256', 'sha512'], default='sha256', help="Hash algorithm to use (default: sha256).")

    args = parser.parse_args()
    file_path = args.file_path
    hash_algorithm = args.algorithm

    checksum = calculate_checksum(file_path, hash_algorithm)
    print(f"{hash_algorithm.upper()} Checksum: {checksum}")

if __name__ == "__main__":
    main()

How it works:

calculate_checksum(file_path, hash_algorithm='sha256'):

* Opens the specified file in binary mode and reads it in chunks.
* Updates the hash function with each chunk until the entire file is processed.
* Returns the computed checksum in hexadecimal format.

main():

* Uses argparse to parse command-line arguments:
* file_path: Path to the file that needs to be checksummed.
* --algorithm: (Optional) Hash algorithm to use (default is SHA-256).
* Computes and prints the checksum using the selected algorithm.

```python
import hashlib
import argparse
import sys

def calculate_checksum(file_path, hash_algorithm='sha256'):
    try:
        hash_func = hashlib.new(hash_algorithm)
        with open(file_path, 'rb') as file:
            while chunk := file.read(8192):
                hash_func.update(chunk)
        return hash_func.hexdigest()
    except FileNotFoundError:
        return f"Error: File '{file_path}' not found."
    except ValueError:
        return f"Error: Invalid hash algorithm '{hash_algorithm}'."
    except Exception as e:
        return f"Error: {str(e)}"

# Step 1: Define the file paths (modify these according to your file locations)
x_train_path = '/content/X_Train_Data_Input.csv'
y_train_path = '/content/Y_Train_Data_Target.csv'

# Step 2: Calculate and print checksums
x_train_checksum = calculate_checksum(x_train_path, 'sha256')
y_train_checksum = calculate_checksum(y_train_path, 'sha256')

print(f"X_Train_Data_Input.csv -> SHA 256: {x_train_checksum}")
print(f"Y_Train_Data_Target.csv -> SHA 256: {y_train_checksum}")

# Step 1: Define the file paths (modify these according to your file locations)
x_test_path = '/content/X_Test_Data_Input.csv'
y_test_path = '/content/Y_Test_Data_Target.csv'

# Step 2: Calculate and print checksums
x_test_checksum = calculate_checksum(x_test_path, 'sha256')
y_test_checksum = calculate_checksum(y_test_path, 'sha256')

print(f"X_Test_Data_Input.csv -> SHA 256: {x_test_checksum}")
print(f"Y_Test_Data_Target.csv -> SHA 256: {y_test_checksum}")
```
